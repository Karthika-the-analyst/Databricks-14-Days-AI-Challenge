```python
Oct_events = spark.read.csv("/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv" , header=True , inferSchema=True )
```

## Basic Operation Using Python

Select (Choose columns)

```python
Oct_events.select(
    "event_time", "event_type", "category_code", "price"
).show(5)
```

Filter (Apply conditions)

```python
Oct_events.filter(
    Oct_events.price > 100
).show(5)
```

GroupBy (Aggregation)

```python
Oct_events.groupBy("event_type").count().show()
```

OrderBy (Sorting)

```python
Oct_events.groupBy("event_type") \
          .count() \
          .orderBy("count", ascending=False) \
          .show()
```

## Basic Operation Using SQL

## Create a TEMP VIEW

You must register the DataFrame as a SQL view.

```python
Oct_events.createOrReplaceTempView("oct_events")
```

Select (Choose columns)

```python
%sql
SELECT event_time, event_type, category_code, price
FROM oct_events
LIMIT 5;
```

Filter (Apply conditions)

```python
%sql
SELECT *
FROM oct_events
WHERE price > 100
LIMIT 5;
```

GroupBy (Aggregation)

```python
%sql
SELECT event_type, COUNT(*) AS total_events
FROM oct_events
GROUP BY event_type;

```

OrderBy (Sorting)

```python
%sql
SELECT event_type, COUNT(*) AS total_events
FROM oct_events
GROUP BY event_type
ORDER BY total_events DESC;
```
