## Task 2: Create Delta Tables (SQL & PySpark)

## Create Delta Table using SQL

Create table from Delta location

```python
%sql
CREATE TABLE IF NOT EXISTS workspace.ecommerce.oct_events_delta_sql
USING DELTA
AS SELECT * FROM delta.`/Volumes/workspace/ecommerce/ecommerce_data/oct_events_delta`;
```

Validate the table

```python
%sql
SELECT COUNT(*) 
FROM workspace.ecommerce.oct_events_delta_sql;
```

## Task 4: Handle Duplicate Inserts

Problem: Duplicate Append

```python
oct_df = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
    header=True,
    inferSchema=True
)

oct_df.write.format("delta") \
    .mode("append") \
    .save("/Volumes/workspace/ecommerce/ecommerce_data/oct_events_delta")
```

Solution: Use MERGE (Upsert)

```python
#Create temp view

oct_df.createOrReplaceTempView("oct_events_temp")
```

```python
%sql
MERGE INTO workspace.ecommerce.oct_events_delta_sql target
USING oct_events_temp source
ON target.event_time = source.event_time
   AND target.user_id = source.user_id
   AND target.product_id = source.product_id
WHEN NOT MATCHED THEN
  INSERT *
```
