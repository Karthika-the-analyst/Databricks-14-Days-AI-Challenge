##Task 2: Write Analytical Queries

Write these queries inside dashboard in add data section

🔹 Revenue Trend (Time Series)

✅ SQL Query
SELECT
  order_date,
  total_revenue
FROM ecommerce_catalog.gold.daily_sales
ORDER BY order_date;


🔹 Orders Trend

✅ SQL Query

SELECT
  order_date,
  total_orders
FROM ecommerce_catalog.gold.daily_sales
ORDER BY order_date;


🔹  Average Order Value (AOV)

✅ SQL Query

SELECT
  order_date,
  avg_order_value
FROM ecommerce_catalog.gold.daily_sales
ORDER BY order_date;


🔹 Funnel Analysis

(Using Silver table – real analytics)

✅ SQL Query

SELECT
  event_type,
  COUNT(DISTINCT user_id) AS users
FROM ecommerce_catalog.silver.orders_clean
GROUP BY event_type
ORDER BY users DESC;


🔹  Top Products by Revenue

✅ SQL Query

SELECT
  product_id,
  SUM(price) AS revenue
FROM ecommerce_catalog.silver.orders_clean
WHERE event_type = 'purchase'
GROUP BY product_id
ORDER BY revenue DESC
LIMIT 10;
