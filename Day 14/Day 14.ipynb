## Task 3: Build a Simple NLP Task

Sample Text Data

```python
reviews = [
    "Sales increased significantly during weekends",
    "Revenue dropped compared to last month",
    "Average order value remained stable"
]
```

Simple NLP Sentiment Logic

```python
def simple_sentiment(text):
    text = text.lower()
    if "increase" in text or "high" in text:
        return "Positive"
    elif "drop" in text or "decrease" in text:
        return "Negative"
    else:
        return "Neutral"

for r in reviews:
    print(r, "->", simple_sentiment(r))
```
